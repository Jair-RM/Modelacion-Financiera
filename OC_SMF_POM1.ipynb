```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun1(x):
  z = np.zeros(2)
  z[0] = x[0]**2 + x[1]**2
  z[1] = (x[0]-5)**2 + (x[1]-5)**2
  return z
```

```python
n = 10**3
z = np.zeros((n,2))
for i in range(n):
  x = np.random.uniform(-5,5,2)
  rs = fun1(x)
  z[i,0] = rs[0]
  z[i,1] = rs[1]

plt.plot(z[:,0],z[:,1],'.')
plt.plot(0,50,'sr')
plt.plot(2,32,'sr')
plt.plot(25,5,'sr')
#plt.plot(200,50,'sr')
plt.show()
```
